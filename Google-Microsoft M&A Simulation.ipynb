DCF Modelling

```python
fcf_estimates = [60000, 65000, 70000, 74000, 78000]
```

```python
growth_rate = 0.025 
wacc = 0.08 
```

```python
terminal_value = fcf_estimates[-1] * (1 + growth_rate) / (wacc - growth_rate)
npv = sum([fcf_estimates[i] / (1 + wacc)**(i + 1) for i in range(5)]) + terminal_value / (1 + wacc)**5
```

```python
npv
```

```python
wacc_range = np.arange(0.07, 0.10, 0.005)
growth_range = np.arange(0.015, 0.035, 0.005)
```

```python
last_fcf = fcf_estimates[-1]
```

```python
sensitivity_table = pd.DataFrame(index=[f"{g*100:.1f}%" for g in growth_range],
                                 columns=[f"{w*100:.1f}%" for w in wacc_range])
```

```python
for g in growth_range:
    for w in wacc_range:
        terminal_val = last_fcf * (1 + g) / (w - g)
        npv_val = sum([fcf_estimates[i] / (1 + w)**(i + 1) for i in range(5)]) + terminal_val / (1 + w)**5
        sensitivity_table.loc[f"{g*100:.1f}%", f"{w*100:.1f}%"] = round(npv_val / 1e3, 1)  # in billions
```

```python
net_debt = -60000 
shares_outstanding = 7450
```

```python
enterprise_value = npv  # previously calculated
equity_value = enterprise_value + net_debt
price_per_share = equity_value * 1e6 / shares_outstanding
```

```python
market_cap_current = 2.8e12
premium = ((equity_value * 1e6) - market_cap_current) / market_cap_current * 100
```

```python
sensitivity_table, round(price_per_share, 2), round(premium, 2)
```

```python
google_eps = 6.00
google_shares = 13_100  # in millions
msft_net_income = 95_000  # in millions
revenue_synergy = 15_000
revenue_synergy_margin = 0.35
cost_synergy = 8_000
```

```python
revenue_synergy_ebit = revenue_synergy * revenue_synergy_margin
total_synergy = revenue_synergy_ebit + cost_synergy
synergies_post_tax = total_synergy * 0.79
```

```python
combined_earnings = msft_net_income + synergies_post_tax
```

```python
msft_shares = 7450  # in millions
exchange_ratio = 1.2
new_google_shares = google_shares + msft_shares * exchange_ratio
```

```python
new_eps = combined_earnings / new_google_shares
eps_change_pct = ((new_eps - google_eps) / google_eps) * 100
```

```python
round(new_eps, 2), round(eps_change_pct, 2)
```

Accretion/Dilution Model Summary
Metric	                          Value
Google EPS (pre-deal)	          $6.00
Pro-forma EPS (post-deal)	      $4.79
EPS Change	                     –20.25%
Conclusion	                      Dilutive

Interpretation:
Despite ~$13.25B in estimated annual synergies (pre-tax), the deal is ~20% dilutive to Google's earnings per share.

Why?
The sheer size of Microsoft makes it hard to absorb without issuing many new shares.
A high exchange ratio (1.2:1 assumed) significantly expands the share count.
Synergies, while large, aren’t enough to fully offset dilution.

New scenario: Cash + Stock Deal
Google pays 60% in stock, 40% in cash (reduces share dilution)

```python
stock_ratio = 0.6
cash_ratio = 0.4
```

```python
new_shares_issued = msft_shares * exchange_ratio * stock_ratio
new_google_shares_cashtock = google_shares + new_shares_issued
```

```python
new_eps_cashtock = combined_earnings / new_google_shares_cashtock
eps_change_pct_cashtock = ((new_eps_cashtock - google_eps) / google_eps) * 100
```

```python
exchange_ratios = [1.0, 1.1, 1.2, 1.3]
synergy_impacts = [10_000, 12_000, 13_250, 15_000]
```

```python
eps_matrix = []
for synergy in synergy_impacts:
    eps_row = []
    for ratio in exchange_ratios:
        shares = google_shares + msft_shares * ratio * stock_ratio
        earnings = msft_net_income + synergy
        eps = earnings / shares
        eps_row.append(round(eps, 2))
    eps_matrix.append(eps_row)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
cax = ax.imshow(eps_matrix, cmap="YlGnBu", aspect='auto')

ax.set_xticks(range(len(exchange_ratios)))
ax.set_xticklabels(exchange_ratios)
ax.set_yticks(range(len(synergy_impacts)))
ax.set_yticklabels([f"${x/1e3:.1f}B" for x in synergy_impacts])
ax.set_xlabel("Exchange Ratio (MSFT : GOOGL)")
ax.set_ylabel("Post-Tax Synergies")

for i in range(len(synergy_impacts)):
    for j in range(len(exchange_ratios)):
        text = ax.text(j, i, f"{eps_matrix[i][j]}", ha="center", va="center", color="black")

plt.title("Pro-Forma EPS under Synergy & Exchange Ratio Scenarios")
fig.colorbar(cax, label="EPS ($)")
plt.tight_layout()
plt.show()

# Return updated EPS and accretion result for cash+stock
round(new_eps_cashtock, 2), round(eps_change_pct_cashtock, 2)
```
